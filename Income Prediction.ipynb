## Dataset Explanation


1. Age :
The age of the individual in years.
2. Workclass :
The type of employment or work classification of the individual. Examples include State-gov, Self-emp-not-inc, Private, etc.
3. Final-weight :
 A weight assigned to the individual for statistical purposes, often used in survey data to represent the individual's contribution to the overall population.
4. Education :
 The highest level of education attained by the individual. Examples include Bachelors, HS-grad, Masters, etc.
5. Education-num :
Numeric representation of the education level, where higher values correspond to higher levels of education. For example, 13 might correspond to Bachelors.
6. Marital-status :
The marital status of the individual. Examples include Never-married, Married-civ-spouse, Divorced, etc.
7. Occupation :
The type of job or occupation of the individual. Examples include Adm-clerical, Exec-managerial, Prof-specialty, etc.
8. Relationship :
The individual's relationship status within the family. Examples include Husband, Wife, Not-in-family, etc.
9. Race :
The race or ethnicity of the individual. Examples include White, Black, Asian-Pac-Islander, etc.
10. Sex :
The gender of the individual. Values are typically Male or Female.
11. Capital-gain :
The amount of capital gain the individual has earned. This is income from investments such as stocks or property.
12. Capital-loss :
The amount of capital loss the individual has incurred. This represents the loss from investments.
13. Weekly-working-hrs :
The number of hours worked per week by the individual.
14. Native-country :
The country where the individual was born. Examples include United-States, Cuba, Jamaica, etc.
15. Income :
The income level of the individual. Values are typically <=50K or >50K, indicating whether the individual earns more or less than $50,000 per year.

### My dataset had null values in the workclass, occupation and native-country which were not displaying as NaNs when I read the data on the notebook, hence replaced the 

### empty strings or special characters with NaN

```python

income_df['Workclass'] = income_df['Workclass'].replace(['', ' ', 'NaN'], np.nan)
income_df['Occupation'] = income_df['Occupation'].replace(['', ' ', 'NaN'], np.nan)
income_df['Native-country'] = income_df['Native-country'].replace(['', ' ', 'NaN'], np.nan)

# Re-checking null values

income_df.isnull().sum()
```

## Data Cleaning



```python
mode_value = income_df['Weekly-working-hrs'].mode()[0]

# Filling in missing values with the mode value
income_df['Weekly-working-hrs'] = income_df['Weekly-working-hrs'].fillna(mode_value)
```

#### Filled in the missing values for capital gain and capital loss with zeros because I assumed the empty rows might imply no capital gain/loss.


```python
income_df['Capital-gain']=income_df['Capital-gain'].fillna(0)
income_df['Capital-loss']=income_df['Capital-loss'].fillna(0)
```

```python
def determine_income(row):
    # Handling missing values in 'Income'
    if pd.isna(row['Income']):
        return '<=50K' if row['Weekly-working-hrs'] > 40 else '>50K'
    return row['Income']

# Applying the function to the DataFrame
income_df['Income'] = income_df.apply(determine_income, axis=1)

income_df.head()
```

```python
income_df.isnull().any()
```

```python
mode_workclass = income_df['Workclass'].mode()[0]
mode_occupation = income_df['Occupation'].mode()[0]
mode_education = income_df['Education'].mode()[0]


income_df['Workclass'] = income_df['Workclass'].fillna(mode_workclass)
income_df['Occupation'] = income_df['Occupation'].fillna(mode_occupation)
income_df['Education'] = income_df['Occupation'].fillna(mode_education)
```

```python
def impute_race(row):
    if pd.isna(row['Race']):
        similar_rows = income_df[(income_df['Education'] == row['Education']) & (income_df['Workclass'] == row['Workclass'])]
        if not similar_rows.empty:
            return similar_rows['Race'].mode()[0]
    return row['Race']

income_df['Race'] = income_df.apply(impute_race, axis=1)
print(income_df)
```

## Label Encoding
#### To convert categorical values, in this case sex into numerical values. Each unique category is assigned a distinct integer label. 

Pros:

Simple and Efficient: Label encoding is computationally efficient and easy to apply.
Maintains Original Data Structure: It maintains the original categorical structure but as integers, which is useful for some algorithms.

Cons:

Ordinal Assumptions: Label encoding introduces an ordinal relationship between categories, which can mislead models that dont expect or require such relationships (e.g., algorithms that do not inherently understand ordinal data).
Not Suitable for All Models: Many machine learning algorithms, especially those that rely on distance metrics (like k-nearest neighbors), might interpret encoded labels incorrectly.


```python
income_df['Sex'] = income_df['Sex'].str.strip().str.title()

# Inspecting unique values
print("Unique values in 'sex':", income_df['Sex'].unique())

# Updating mapping dictionary with all unique values
sex_mapping = {'Male': 1, 'Female': 0, 'Other': 2, 'Unknown': 3}

# Applying the mapping
income_df['Sex-encoded'] = income_df['Sex'].map(sex_mapping)

# Displaying the processed DataFrame
print(income_df)
```

## Frequency Encoding

Pros:

Simple Implementation: Frequency encoding is straightforward and easy to implement.
Preserves Information: Unlike some encoding methods, frequency encoding preserves the distribution of the categorical variables, maintaining information about the prevalence of each category.
Reduces Dimensionality: Converts categorical variables into numerical format without increasing the number of features.

Cons:

Assumption of Ordinality: Frequency encoding implicitly assumes an ordinal relationship between categories, which might not be appropriate for all models or data.
Potential for Misleading Information: High-frequency categories might dominate the model, potentially overshadowing other important features.

```python
frequency_encoding = income_df['Workclass'].value_counts()

# Mapping frequencies to the 'workclass' column
income_df['Workclass-encoded'] = income_df['Workclass'].map(frequency_encoding)

print(income_df)
```

## Reducing Cardinality
#### Reducing cardinality for Native-country helps simplify the model and makes it easier to interpret. The threshold ensures that only truly rare categories are combined.
#### To combine less frequent categories into an 'Other' category to reduce cardinality and improve model efficiency.

Pros:

Improved Model Performance: Reducing cardinality can help in simplifying the model and improving its performance by decreasing the complexity and overfitting.
Reduced Dimensionality: Fewer categories mean fewer features in your model, which can speed up training time and reduce computational resources.
Enhanced Interpretability: Fewer categories often make the models results easier to interpret and analyze.

Cons:

Loss of Information: Grouping or merging categories can lead to a loss of information, potentially oversimplifying the data and losing important distinctions between categories.
Arbitrary Grouping: The process of combining categories can sometimes be arbitrary and might not always align with the underlying data patterns.

```python
def reduce_cardinality(df, column, threshold=0.01):
    counts = df[column].value_counts(normalize=True)
    rare_categories = counts[counts < threshold].index
    df[column] = df[column].replace(rare_categories, 'Other')
    return df

income_df = reduce_cardinality(income_df, 'Native-country', 0.01)


print(income_df.head(10))
```

```python
def reduce_occupation_cardinality(df, threshold=0.01):
    counts = df['Occupation'].value_counts(normalize=True)
    rare_categories = counts[counts < threshold].index
    df['Occupation'] = df['Occupation'].replace(rare_categories, 'Other')
    return df

income_df = reduce_occupation_cardinality(income_df, 0.01)
```
